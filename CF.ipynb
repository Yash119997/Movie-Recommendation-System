```python
# assign weights
from numpy import sqrt 
def ass_w (user, other, data):
    sum = 0
    sum_u = 0
    sum_ou = 0
    n = 0
    rating1 = data[user]
    rating2 = data[other]
    leng = len(rating1)
    
    m1 = training1['mean_rating'][user]
    m2 = training1['mean_rating'][other]
    
    for i in range(leng):
        if rating1[i]>0: 
            if rating2[i]>0:
                n += 1
                sum += (rating1[i]-m1)*(rating2[i]-m2)
    
    if n == 0:
        return 0
    else:

        for value in rating1[:]:
            sum_u += value**2
        for value in rating2[:]:
            sum_ou += value**2

            
        f1 = sum_u - m1
        f2 = sum_ou - m2
        denominator = sqrt(sum_u) * sqrt(sum_ou)

        if denominator == 0:
            return 0
        else:
            return sum/denominator

ass_w(1,2,train_data_matrix)
```

```python
# create weight matrix
n = len(train_data_matrix)
w = np.empty(shape=(n,n),dtype='object')
for i in range(n):
    for j in range(n):
        w[i][j] = 0

for i in range(n):
    for j in range(n):
        w[i][j] = ass_w(i,j,train_data_matrix)
```

```python
# weights using pearson correlation
from sklearn.metrics.pairwise import pairwise_distances
user_similarity = 1 - pairwise_distances(train_data_matrix, metric='correlation')
```

```python
def predictions(data, user_sim):
    user_mean = data.mean(axis=1)
    difference = (data - user_mean[:, np.newaxis])
    pred = user_mean[:, np.newaxis] + user_sim.dot(difference) / np.array([np.abs(user_sim).sum(axis=1)]).T
    return pred
```

```python
user_prediction = predictions(train_data_matrix, user_similarity)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
def err(pred_value, true_val):
    pred_value = pred_value[true_val.nonzero()].flatten()
    true_val = true_val[true_val.nonzero()].flatten()
    return sqrt(mean_squared_error(pred_value, true_val))
```
